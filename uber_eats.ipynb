## <div style="padding: 20px;color:white;margin:10;font-size:90%;text-align:left;display:fill;border-radius:10px;overflow:hidden;background-image: url(https://w0.peakpx.com/wallpaper/957/661/HD-wallpaper-white-marble-white-stone-texture-marble-stone-background-white-stone.jpg)"><b><span style='color:black'> Most common restaurants</span></b> </div>

```python
#Investigate the restaurants which have alot of items on their menus. 
conn_count = duckdb.connect('uber_restaurants.db')
query = """
SELECT restaurant_id, COUNT(DISTINCT category) AS category_count
FROM restaurants
GROUP BY restaurant_id
ORDER BY category_count DESC
LIMIT 15;
"""
results_count = conn_count.execute(query).fetchdf()
df_ubereats_count = pd.DataFrame(results_count) 
display(df_ubereats_count)
conn_count.close()
```

## <div style="padding: 20px;color:white;margin:10;font-size:90%;text-align:left;display:fill;border-radius:10px;overflow:hidden;background-image: url(https://w0.peakpx.com/wallpaper/957/661/HD-wallpaper-white-marble-white-stone-texture-marble-stone-background-white-stone.jpg)"><b><span style='color:black'> Items price ranges</span></b> </div>

```python
#Investigate the price ranges of items for the restaurant with alot of items. 
conn_pricerange = duckdb.connect('uber_restaurants.db')
query_pricerange = """
SELECT category, 
       MIN(CAST(SUBSTR(price, 1, LENGTH(price) - 4) AS FLOAT)) AS min_price, 
       MAX(CAST(SUBSTR(price, 1, LENGTH(price) - 4) AS FLOAT)) AS max_price
FROM restaurants
WHERE restaurant_id = 31582
GROUP BY category
LIMIT 20;
"""
results_pricerange = conn_pricerange.execute(query_pricerange).fetchdf()
df_ubereats_pricerange = pd.DataFrame(results_pricerange) 
display(df_ubereats_pricerange)
conn_pricerange.close()
```

## <div style="padding: 20px;color:white;margin:10;font-size:90%;text-align:left;display:fill;border-radius:10px;overflow:hidden;background-image: url(https://w0.peakpx.com/wallpaper/957/661/HD-wallpaper-white-marble-white-stone-texture-marble-stone-background-white-stone.jpg)"><b><span style='color:black'> Most expensive restaurants</span></b> </div>

```python
#Investigate the top 15 restaurants which have high average prices of items. 
conn_avgprice = duckdb.connect('uber_restaurants.db')
query_avgprice = """
    WITH top_restaurants AS (
    SELECT restaurant_id, COUNT(DISTINCT category) AS category_count
    FROM restaurants
    GROUP BY restaurant_id
    ORDER BY category_count DESC
    LIMIT 15
)
SELECT r.restaurant_id, AVG(CAST(SUBSTR(r.price, 1, LENGTH(r.price) - 4) AS FLOAT)) AS avg_price
FROM restaurants r
JOIN top_restaurants tr ON r.restaurant_id = tr.restaurant_id
GROUP BY r.restaurant_id
ORDER BY avg_price DESC
"""
results_avgprice = conn_avgprice.execute(query_avgprice).fetchdf()
df_ubereats_avgprice = pd.DataFrame(results_avgprice) 
display(df_ubereats_avgprice)
conn_avgprice.close()
```

## <div style="padding: 20px;color:white;margin:10;font-size:90%;text-align:left;display:fill;border-radius:10px;overflow:hidden;background-image: url(https://w0.peakpx.com/wallpaper/957/661/HD-wallpaper-white-marble-white-stone-texture-marble-stone-background-white-stone.jpg)"><b><span style='color:black'> Snacks price analysis</span></b> </div>

```python
#Investigate restaurants which have the most expensive snacks.
conn_snacks = duckdb.connect('uber_restaurants.db')
query_snacks = '''
SELECT restaurant_id, AVG(CAST(SUBSTR(price, 1, LENGTH(price) - 4) AS FLOAT)) AS avg_price
FROM restaurants 
WHERE category = 'Snacks' 
GROUP BY restaurant_id
ORDER BY avg_price DESC 
LIMIT 10;
'''
results_snacks = conn_snacks.execute(query_snacks).fetchdf()
df_ubereats_snacks = pd.DataFrame(results_snacks) 
display(df_ubereats_snacks)
conn_snacks.close()
```

```python
#Investigate the most expensive restaurant
conn_expensive = duckdb.connect('uber_restaurants.db')
query_expensive = '''
SELECT restaurant_id, AVG(CAST(SUBSTR(price, 1, LENGTH(price) - 4) AS FLOAT)) AS avg_price
FROM restaurants  
GROUP BY restaurant_id
ORDER BY avg_price DESC 
LIMIT 10;
'''
results_expensive = conn_expensive.execute(query_expensive).fetchdf()
df_ubereats_expensive = pd.DataFrame(results_expensive) 
display(df_ubereats_expensive)
conn_expensive.close()
```
