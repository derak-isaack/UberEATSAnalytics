```python
from IPython.display import display
import duckdb  
import pandas as pd 
```

```python
#initiualize a duckdb connection
conn = duckdb.connect('uber_restaurants.db')
```

```python
conn2 = duckdb.connect('uber_restaurants.db')
#Read in data from parquet file using fetchall.
results = conn2.execute("CREATE TABLE restaurants AS SELECT * FROM 'uberEats-menus.parquet'").fetchall()
df_ubereats = pd.DataFrame(results)

display(df_ubereats.head(10))

conn2.close()
```

```python
conn3 = duckdb.connect('uber_restaurants.db')
results2 = conn3.execute('SELECT * FROM restaurants LIMIT 10').fetchdf()
df_ubereats2 = pd.DataFrame(results2) 
display(df_ubereats2)
conn3.close()
```
