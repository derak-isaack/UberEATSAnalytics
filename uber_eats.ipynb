```python
#Import necessary libraries for data manipulation and display. 
from IPython.display import display
import duckdb  
import pandas as pd 
```

```python
#initiualize a duckdb connection
conn = duckdb.connect('uber_restaurants.db')
```

```python
conn2 = duckdb.connect('uber_restaurants.db')
#Read in data from parquet file using fetchall.
results = conn2.execute("CREATE TABLE restaurants AS SELECT * FROM 'uberEats-menus.parquet'").fetchall()
df_ubereats = pd.DataFrame(results)

display(df_ubereats.head(10))

conn2.close()
```

```python
conn3 = duckdb.connect('uber_restaurants.db')
results2 = conn3.execute('SELECT * FROM restaurants LIMIT 30').fetchdf()
df_ubereats2 = pd.DataFrame(results2) 
display(df_ubereats2)
conn3.close()
```

```python
#initiualize a duckdb connection for the csv file. 
conn4 = duckdb.connect('uber_restaurants_listings.db')
```

```python
conn5 = duckdb.connect('uber_restaurants_listings.db')
results_rest = conn5.execute("CREATE TABLE restaurants_listings AS SELECT * FROM 'restaurants.csv'").fetchdf()
df_restaurant = pd.DataFrame(results_rest) 
display(df_restaurant)
conn5.close()
```

```python
conn6 = duckdb.connect('uber_restaurants_listings.db')
results_rest2 = conn6.execute('SELECT * FROM restaurants_listings LIMIT 10').fetchdf()
df_ubereats_rest = pd.DataFrame(results_rest2) 
display(df_ubereats_rest)
conn3.close()
```

```python
#Investigate the restaurants which have alot of items on their menus. 
conn_count = duckdb.connect('uber_restaurants.db')
query = """
SELECT restaurant_id, COUNT(DISTINCT category) AS category_count
FROM restaurants
GROUP BY restaurant_id
ORDER BY category_count DESC
LIMIT 15;
"""
results_count = conn_count.execute(query).fetchdf()
df_ubereats_count = pd.DataFrame(results_count) 
display(df_ubereats_count)
conn_count.close()
```
